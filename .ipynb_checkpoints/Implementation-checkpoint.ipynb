```python
image = cv.imread(cv.samples.findFile("Test-data/050200074.jpg"), 0);
(h, w) = np.shape(image);
plt.imshow(image)
width = int (findComponents(image));
```

```python
# replace pixel with avg gray value 
def PPA(image, width):
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 32),
                         sharex=True, sharey=True)

    ax = axes.ravel()
    
    stripped_image, strips = putGLM(np.copy(image), width);
#     plt.imshow(stripped_image);
#     plt.show();
    ax[0].imshow(stripped_image);
    ax[0].axis('off');
    ax[0].set_title('GLM Image', fontsize=20);
    
    # apply otsu's algo
    th = np.copy(stripped_image);
    for i in range(strips):
        l = i * width;
        r = (i+1) * width if (((i+1) * width) < w) else w
        _,th[:,l:r] = cv.threshold(stripped_image[:,l:r],0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
#     plt.imshow(th);
#     plt.show()
    ax[1].imshow(th);
    ax[1].axis('off');
    ax[1].set_title('Otsu Algo', fontsize=20);
    
    filteredWImage = filterWhite(np.copy(th), strips, width)
    Wheight = avgWhiteH(np.copy(th), strips, width)
#     plt.imshow(filteredWImage);
#     plt.show();
    ax[2].imshow(filteredWImage);
    ax[2].axis('off');
    ax[2].set_title('Filtered White Blocks', fontsize=20);
    
    filteredBImage = filterBlack(np.copy(filteredWImage), strips, width)
#     plt.imshow(filteredBImage);
#     plt.show();
    ax[3].imshow(filteredBImage);
    ax[3].axis('off');
    ax[3].set_title('Filtered Black Blocks', fontsize=20);
    
    mode = modeWhite(np.copy(filteredBImage), strips, width);
    return mode, filteredBImage, strips, Wheight;
```

```python
new_width, filteredBImage, strips, height = PPA(image, width)
```

```python
mode, filteredBImage, strips, height = PPA(image, new_width)
```

```python
# Remove black block equal or greater than twice of avg
avgBH = removeBlack(np.copy(filteredBImage), strips, new_width);
plt.imshow(avgBH)
```

```python
# Invert image and then dilate and invert it back
img = [];
for row in avgBH:
    rows = [];
    for ele in row:
        if (ele == 255):
            rows.append(0);
        else:
            rows.append(255);
    img.append(rows);
img = np.array(img, dtype = np.uint8)
```

```python
dilated_img = cv.dilate(img, np.ones((1,4* new_width),np.uint8), iterations=1)
plt.imshow(dilated_img);
```

```python
img = [];
for row in dilated_img:
    rows = [];
    for ele in row:
        if (ele == 255):
            rows.append(0);
        else:
            rows.append(1);
    img.append(rows);
img = np.array(img, dtype = np.uint8);
plt.imshow(img);
```

```python
# Perform thinning

skel = skeletonize(img)
plt.imshow(skel)
```

```python
# Apply Sobel filter
sobel_img = sobel_h(skel)
plt.imshow(sobel_img)
```

```python
def constructLinesH(image, strips, width, height):
    (h, w) = np.shape(image);
    dist = int (height);

    for i in range(1,h-1):
        for j in range(1,w-1):
            if lonelyHStart(image, i, j) == 0:
                image = findLineH(image, i, j, dist);
    plt.imshow(image);
```

```python
def lonelyHStart(image, i, j):
    (h, w) = np.shape(image);
    count = 0;
    if (image[i,j] == 1):
        if (j+1 < w and image[i, j+1] == 1 ):
            count+=1
        if (j-1 > 0 and image[i, j-1] == 1 ):
            count+=1
        if (i+1 < h and image[i+1, j] == 1 ):
            count+=1
        if (i+1 < h and j+1 < w and image[i+1, j+1] == 1 ):
            count+=1
        if (i+1 < h and j-1 > 0 and image[i+1, j-1] == 1 ):
            count+=1
    return count;

def lonelyHEnd(image, i, j):
    (h, w) = np.shape(image);
    count = 0;
    if (image[i,j] == 1):
        if (i-1 > 0 and image[i-1, j] == 1 ):
            count+=1
        if (j+1 < w and image[i, j+1] == 1 ):
            count+=1
        if (j-1 > 0 and image[i, j-1] == 1 ):
            count+=1
        if (i-1 > 0 and j+1 < w and image[i-1, j+1] == 1 ):
            count+=1
        if (i-1 > 0 and j-1 > 0 and image[i-1, j-1] == 1 ):
            count+=1
    return count;

def findLineH(image, i, j, dist):
    (h, w) = np.shape(image);

    for y in range(dist):
        k = i-y if i-y > 0 else 0
        for x in range(j+1,w):
            if (lonelyHEnd(image, k, x) == 0):
                print (str((i, j)) +  str((k, x)));
                cv.line(image, (j, i), (x, k), 1, 1, cv.LINE_AA)
                return image;
        k = i+y if i+y < h else h-1
        for x in range(j+1, w):
            if (lonelyHEnd(image, k, x) == 0):
                print (str((i, j)) +  str((k, x)));
                cv.line(image, (j, i), (x, k), 1, 1, cv.LINE_AA)
                return image;
    if (j < w/2):
        cv.line(image, (0, i), (j, i), 1, 1, cv.LINE_AA)
    else:
        cv.line(image, (w, i), (j, i), 1, 1, cv.LINE_AA)
    return image;
```
